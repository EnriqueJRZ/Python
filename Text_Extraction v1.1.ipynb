# Extracción, visualización y guardado de datos

A continuación se definen las funciones para extraer, visualizar y cargar datos de archivos de texto varios.

```python
# Extracción del contenido de un archivo (de nuestro directorio en cuestión)
def extract_data(file_name):
    with open('./../EXCELSIOR_100_files/' + file_name, encoding='utf-8') as file:
        content = file.read()
        content = content.lower()
        file.close()
    return content
```

```python
# Visualización de la información contenida en un archivo
def visualize_data(content):
    print(content)
```

```python
# Guardado de la información previamente extraída de un determinado archivo 
def save_data(file_name, content):
    with open('./new_EXCELSIOR_100_files/' + file_name, 'w', encoding = 'utf-8') as file:
        file.write(content)
```

```python
# Limpieza del texto quitando etiquetas HTML
from bs4 import BeautifulSoup
def clean_text(text):
    soup = BeautifulSoup(text, 'lxml')
    return soup.get_text()
```

```python
# Leemos cada uno de los archivos del directorio en cuestión
import nltk
words = []
for article_number in range(960401, 960718):
    file_name = 'e' + str(article_number) + '_mod.htm'
    try:
        print('\nArchivo ' + file_name)
        content = extract_data(file_name)
        print('Se extrajo exitosamente el texto')
        #visualize_data(content)
        content = clean_text(content)
        print('Se limpió exitosamente el texto')
        words_aux = nltk.word_tokenize(content)
        #words_aux = content.split()
        for word in words_aux:
            words.append(word)
        print('Se agregaron las palabras extraídas a la lista de palabras')   
        save_data(file_name, content)
        print('Se guardó correctamente el archivo')
    except Exception as e:
        print('Error ', e)
        print('Ocurrió un error al procesar el archivo')
```

```python
# Convertimos la lista de palabras extraídas a vocabulario
v = set(words)
```
